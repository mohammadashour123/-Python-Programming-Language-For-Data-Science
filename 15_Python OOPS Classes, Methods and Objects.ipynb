```python
class Car : 
    def __init__(self ,windows , door ,brand ):
        self.windows = windows
        self.door = door 
        self.brand = brand 
    def type_brand(self) : 
        return "this is a brand {} of car".format(self.brand)
```

```python
car1 = Car()
```

```python
car2 = Car(4,4,"bmw")
```

```python
car2.type_brand()
```
