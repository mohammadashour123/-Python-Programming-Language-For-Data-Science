- A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
- Dictionaries are written with curly brackets, and have keys and values:

```python
computer_science = {

    "WEB" : {
       "front end" : ["html" , "css" , "java script" , "bootstap" , "react" , "angular" , "responsive web"] ,
       "back end" : ["html" , "css" , "php" , "database" , "sass"] 

    } ,
    "data science" : {
        "Data analysis" : ["statistics" , "python" , "R" , "ml" , "DL" , "AI "] , 
        "Data engneer" : ["database" , "sql" , "mongodb" , "ml" , "DL" , "AI "] , 

    } , 

    "Mobile application" : {
        "flutter" : ["Dart" , "responsive" , "flutter" , "wedget"], 
        "android" : ["java" , "kotlen" , "oop" , "responsive"]
    }

}
```

```python
print(computer_science)
```

```python
print(computer_science["data science"]["Data analysis"]) # ['statistics', 'python', 'R', 'ml', 'DL', 'AI ']
```

```python
print(computer_science["data science"]["Data engneer"]) # ['database', 'sql', 'mongodb', 'ml', 'DL', 'AI ']
```

```python
print(computer_science["WEB"]["front end"]) # ['html', 'css', 'java script', 'bootstap', 'react', 'angular', 'responsive web']
```

```python
print(computer_science["Mobile application"]["flutter"][1]) # responsive
```

```python
print(len(computer_science)) # هيطبع 3 عشان عندنا 3 قواميس جوا 
```
